```python
import pandas as pd

data=pd.DataFrame({'ProductName':['phone','earphone','kurti','Fridge','Dishwasher'],
                   'Category':['electronic','electronic','cloths','kitchen','electronic'],
                   'price':[20000,50000,30000,6000,89000],
                   'Quantity_sold':[2,3,4,7,6]})
data
```

```python
p = data.pivot_table(index=['Category'],values=['price'],aggfunc='mean')
p
```

```python
data['revenue']=data['price']*data['Quantity_sold']

p = data.pivot_table(index=['Category'],values=['revenue','price'],aggfunc='mean')
p
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("mtcars.csv")
df
```

```python
#Histogram
plt.title("HISTOGRAM")
hist=sns.histplot(x='mpg',bins=10,data=df)
hist
```

```python
plt.title("SCATTER PLOT")
scatter=sns.scatterplot(x='wt',y='mpg',data=df,hue='wt')
scatter
```

```python
plt.title("COUNTPLOT")
bar=sns.countplot(x='am',data=df)
bar
```

```python
plt.title("BOX PLOT")
box=sns.boxplot(x='mpg',data=df)
box
```
